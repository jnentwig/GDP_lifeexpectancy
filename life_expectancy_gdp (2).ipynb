```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv('all_data.csv')
print(data.Country.unique())
data.rename(columns = {'Life expectancy at birth (years)': 'Life_expectancy'}, inplace=True)
data['GDP_change'] = data['GDP'].pct_change()
data['life_change'] = data['Life_expectancy'].pct_change()
print(list(data.columns))
#years 2000-2015
```

```python
#make a function to calculate change in GDP
data['GDP_change'] = data['GDP'].pct_change()
data['life_change'] = data['Life_expectancy'].pct_change()
plt.subplot(2,1,1)
sns.scatterplot(x = 'Year', y = 'GDP_change', data = data[data.Year != 2000], hue = 'Country')
plt.legend(loc = 'upper left', fontsize='8')
plt.subplot(2,2)
sns.scatterplot(x = 'Year', y = 'GDP', data = data, hue = 'Country')
plt.show()
plt.clf()
sns.scatterplot(x = 'Year', y = 'life_change', data = data[data.Year != 2000], hue = 'Country')
plt.show()
```

```python
sns.scatterplot(x = 'Year', y = 'GDP', data = data, hue = 'Country')
plt.legend(loc = 'upper left', fontsize='8')
plt.show()
plt.clf()
sns.scatterplot(x = 'Year', y = 'Life_expectancy', data = data, hue = 'Country')
plt.show()
```

```python
#plot and predict GDP vs life expectancy
sns.scatterplot(x = 'GDP', y='Life_expectancy', data=data[data.Country!= 'Zimbabwe'], hue='Country')
plt.show()
plt.clf()
#which countrys follow or dont follow the trend?
plt.subplot(1,2,1)
plt.plot( data.Year[data.Country == 'United States of America'],data.GDP[data.Country == 'United States of America'])
plt.subplot(1,2,2)
plt.plot(data.Year[data.Country == 'China'],data.GDP[data.Country == 'China'])
plt.tight_layout()
```
